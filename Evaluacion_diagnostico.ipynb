### Cargando los datos

Para cargar los datos ocuparemos la libreria Pandas

```python
import pandas as pd

tabla_datos = pd.read_json('farmers-protest-tweets-2021-03-5.json', lines=True)
tabla_datos.head()
```

### Consultas
####  1. Los top 10 tweets más retweeted

Para realizar de forma mas fácil la función, se utilizó la documentación de ocupe la documentacion: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html

```python
def TopRetweets(tabla):
    return tabla.sort_values(by="retweetCount", ascending = False)[["user", "content", "retweetCount", "date"]][0:10]

TopRetweets(tabla_datos)
```

####  2. Los top 10 usuarios en función a la cantidad de tweets que emitieron

Se utilizaron los contenidos de IIC2233 Programación Avanzada, las librerías "json" y "collections"

```python
import json
from collections import defaultdict

def TopTweetsUsuarios(tabla):
    usuarios = defaultdict(int)
    for tweet in range(len(tabla)):
        usuarios[tabla.iloc[tweet, 5]['username']] += 1
    usuarios = sorted(usuarios.items(), key=lambda x: x[1], reverse = True)
    return pd.DataFrame(usuarios).rename(columns = {0: 'nombre_usuario', 1: 'numero_tweets'}, inplace = False)[0:10]

TopTweetsUsuarios(tabla_datos)
```
